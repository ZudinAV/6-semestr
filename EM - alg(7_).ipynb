```python
import numpy as np
from scipy.stats import binom
```

```python
M = 10
y = np.array([ 7, 7, 1, 3, 2, 1, 5, 1, 4, 0,  2, 3, 3, 1, 2, 2, 3, 1, 4, 0,  8, 3, 5, 3, 0, 0, 7, 1, 1, 3,  1, 3, 2, 4, 1, 6, 2, 2, 4, 1,  3, 1, 1, 2, 7, 3, 3, 2, 2, 2 ])
```

```python
np.random.seed(42)
μ = np.random.rand(2)
π = np.array([0.5, 0.5])
π_1, μ_1 = None, None
eps = 1e-8
```

```python
ep = 0
while True:
    print(f'Step {ep}:')
    print(f'π = {π}')
    print(f'μ = {μ}')
    print()
    ep += 1
    r = binom.pmf(y[:, None], M, μ[None, :]) * π[None, :]
    r = r / np.sum(r, axis=1)[:, None]
    π_1 = π.copy()
    μ_1 = μ.copy()
    π = r.mean(axis=0)
    μ = np.sum(y[:, None] * r, axis=0) / r.sum(axis=0) / M
    if np.abs(π - π_1).sum() < eps and np.abs(μ - μ_1).sum() < eps:
        break
```

```python
print(f'π = {π}') #Результат
print(f'μ = {μ}') #Результат
```
